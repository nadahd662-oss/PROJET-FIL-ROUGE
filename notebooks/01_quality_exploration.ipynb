```python
# Check unique values in categorical columns that are likely inconsistent
categorical_cols = ['gender', 'grade_level', 'status', 'internet_access', 'parent_education', 'scholarship']

print("=== STUDENTS - Valeurs catégorielles uniques ===\n")
for col in categorical_cols:
    print(f"--- {col} ---")
    print(df_students[col].value_counts(dropna=False))
    print()
```

```python
# Columns that should be numeric but stored as text
suspect_cols = ['age', 'math_score', 'attendance_pct', 'average_score']

print("=== Valeurs non-numériques cachées ===\n")
for col in suspect_cols:
    # Try to convert to number — anything that fails becomes NaN
    converted = pd.to_numeric(df_students[col], errors='coerce')
    # Keep only values that failed conversion but were not already NaN
    weird_values = df_students[col][converted.isna() & df_students[col].notna()].unique()
    print(f"--- {col} ---")
    print(weird_values)
    print()
```

```python
# Force convert to numeric for outlier inspection
num_cols = ['age', 'math_score', 'attendance_pct', 'average_score', 'study_hours_per_day']

print("=== Statistiques descriptives (colonnes numériques) ===\n")
for col in num_cols:
    converted = pd.to_numeric(df_students[col], errors='coerce')
    print(f"--- {col} ---")
    print(converted.describe())
    print(f"Valeurs négatives : {(converted < 0).sum()}")
    if col == 'age':
        print(f"Valeurs > 25 : {(converted > 25).sum()}")
    if col in ['math_score', 'average_score']:
        print(f"Valeurs > 20 : {(converted > 20).sum()}")
    if col == 'attendance_pct':
        print(f"Valeurs > 100 : {(converted > 100).sum()}")
    if col == 'study_hours_per_day':
        print(f"Valeurs > 16 : {(converted > 16).sum()}")
    print()
```

```python
# Count how many dates follow each format
format_counts = {
    'YYYY-MM-DD (ex: 2023-09-01)' : 0,
    'DD/MM/YYYY (ex: 01/09/2023)' : 0,
    'DD Month YYYY (ex: 1 September 2023)' : 0,
    'Autre / Inconnu' : 0
}

for date in df_students['enrollment_date']:
    date = str(date).strip()
    if '-' in date and date[:4].isdigit():
        format_counts['YYYY-MM-DD (ex: 2023-09-01)'] += 1
    elif '/' in date:
        format_counts['DD/MM/YYYY (ex: 01/09/2023)'] += 1
    elif any(month in date for month in ['January','February','March','April','May','June',
                                          'July','August','September','October','November','December']):
        format_counts['DD Month YYYY (ex: 1 September 2023)'] += 1
    else:
        format_counts['Autre / Inconnu'] += 1

print("\n=== Formats détectés ===\n")
for format_name, count in format_counts.items():
    print(f"{format_name} : {count} valeurs")
```
