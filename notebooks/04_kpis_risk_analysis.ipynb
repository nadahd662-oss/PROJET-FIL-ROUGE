```python
import pandas as pd

# Load silver data
df_students = pd.read_csv("../data/2_silver/students_silver.csv")
df_schools = pd.read_csv("../data/2_silver/schools_silver.csv")
df_regions = pd.read_csv("../data/2_silver/regional_stats_silver.csv")

# Join all 3 sources into one master table
df = df_students.merge(df_schools, on='school_id', how='left')
df = df.merge(df_regions, on='region', how='left')

print("✅ Data loaded and joined")
print(f"Shape : {df.shape}")
```

```python
import os

# Create gold folder if it doesn't exist
os.makedirs("../data/3_gold/", exist_ok=True)

# Save master joined table
df.to_csv("../data/3_gold/gold_master.csv", index=False)

print("✅ Gold master saved")
print(f"Shape : {df.shape}")
```

```python
# Count students by status
total = len(df)
admis = len(df[df['status'] == 'Admis'])
ajourne = len(df[df['status'] == 'Ajourné'])
recale = len(df[df['status'] == 'Recalé'])

# Build KPI table
kpi_global = pd.DataFrame({
    'status': ['Admis', 'Ajourné', 'Recalé'],
    'nombre_eleves': [admis, ajourne, recale],
    'taux_pct': [round(admis/total*100, 2),
                 round(ajourne/total*100, 2),
                 round(recale/total*100, 2)]
})

print("=== KPI 1 : Taux de réussite global ===\n")
print(kpi_global)
```

```python
# Count total and admis per region
total_region = df.groupby('region')['status'].count()
admis_region = df[df['status'] == 'Admis'].groupby('region')['status'].count()

# Build KPI table
kpi_region = pd.DataFrame({
    'region': total_region.index,
    'total_eleves': total_region.values,
    'admis': admis_region.values,
    'taux_reussite_pct': round(admis_region / total_region * 100, 2).values
}).sort_values('taux_reussite_pct', ascending=False).reset_index(drop=True)

print("=== KPI 2 : Taux de réussite par région ===\n")
print(kpi_region)
```

```python
# By school type
total_type = df.groupby('school_type')['status'].count()
admis_type = df[df['status'] == 'Admis'].groupby('school_type')['status'].count()

kpi_type = pd.DataFrame({
    'school_type': total_type.index,
    'total_eleves': total_type.values,
    'admis': admis_type.values,
    'taux_reussite_pct': round(admis_type / total_type * 100, 2).values
}).reset_index(drop=True)

print("=== KPI 3a : Réussite par type d'établissement ===\n")
print(kpi_type)

# By milieu
total_milieu = df.groupby('milieu')['status'].count()
admis_milieu = df[df['status'] == 'Admis'].groupby('milieu')['status'].count()

kpi_milieu = pd.DataFrame({
    'milieu': total_milieu.index,
    'total_eleves': total_milieu.values,
    'admis': admis_milieu.values,
    'taux_reussite_pct': round(admis_milieu / total_milieu * 100, 2).values
}).reset_index(drop=True)

print("\n=== KPI 3b : Réussite par milieu ===\n")
print(kpi_milieu)
```

```python
# By grade level
total_level = df.groupby('grade_level')['status'].count()
admis_level = df[df['status'] == 'Admis'].groupby('grade_level')['status'].count()

kpi_level = pd.DataFrame({
    'grade_level': total_level.index,
    'total_eleves': total_level.values,
    'admis': admis_level.values,
    'taux_reussite_pct': round(admis_level / total_level * 100, 2).values
}).reset_index(drop=True)

print("=== KPI 4a : Réussite par niveau ===\n")
print(kpi_level)

# By gender
total_gender = df.groupby('gender')['status'].count()
admis_gender = df[df['status'] == 'Admis'].groupby('gender')['status'].count()

kpi_gender = pd.DataFrame({
    'gender': total_gender.index,
    'total_eleves': total_gender.values,
    'admis': admis_gender.values,
    'taux_reussite_pct': round(admis_gender / total_gender * 100, 2).values
}).reset_index(drop=True)

print("\n=== KPI 4b : Réussite par genre ===\n")
print(kpi_gender)

# By internet access
total_internet = df.groupby('internet_access')['status'].count()
admis_internet = df[df['status'] == 'Admis'].groupby('internet_access')['status'].count()

kpi_internet = pd.DataFrame({
    'internet_access': total_internet.index,
    'total_eleves': total_internet.values,
    'admis': admis_internet.values,
    'taux_reussite_pct': round(admis_internet / total_internet * 100, 2).values
}).reset_index(drop=True)

print("\n=== KPI 4c : Réussite par accès internet ===\n")
print(kpi_internet)
```

```python
# Average attendance and study hours by status
kpi_engagement = df.groupby('status')[['attendance_pct', 'study_hours_per_day', 'average_score']].mean().round(2)

print("=== KPI 5 : Engagement par statut ===\n")
print(kpi_engagement)
```

```python
# Calculate risk score for each student
df['risk_score'] = (
    (1 - (df['average_score'] / 20)) * 0.5 +
    (1 - (df['attendance_pct'] / 100)) * 0.3 +
    (1 - (df['study_hours_per_day'] / 7)) * 0.2
).round(2)

# Classify risk level
def classify_risk(score):
    if score >= 0.6:
        return 'Risque élevé'
    elif score >= 0.4:
        return 'Risque modéré'
    else:
        return 'Risque faible'

df['risk_level'] = df['risk_score'].apply(classify_risk)

# Summary table
kpi_risk = df.groupby('risk_level').agg(
    nombre_eleves = ('student_id', 'count'),
    moyenne_score = ('average_score', 'mean'),
    moyenne_assiduite = ('attendance_pct', 'mean'),
    moyenne_heures_etude = ('study_hours_per_day', 'mean')
).round(2)

print("=== KPI 6 : Segmentation par niveau de risque ===\n")
print(kpi_risk)
```

```python
# Save all KPI tables to gold layer
kpi_global.to_csv("../data/3_gold/kpi_global.csv", index=False)
kpi_region.to_csv("../data/3_gold/kpi_region.csv", index=False)
kpi_type.to_csv("../data/3_gold/kpi_type.csv", index=False)
kpi_milieu.to_csv("../data/3_gold/kpi_milieu.csv", index=False)
kpi_level.to_csv("../data/3_gold/kpi_level.csv", index=False)
kpi_gender.to_csv("../data/3_gold/kpi_gender.csv", index=False)
kpi_internet.to_csv("../data/3_gold/kpi_internet.csv", index=False)
kpi_engagement.to_csv("../data/3_gold/kpi_engagement.csv")
kpi_risk.to_csv("../data/3_gold/kpi_risk.csv")

print("✅ Tous les KPIs sauvegardés dans data/3_gold/")
```
